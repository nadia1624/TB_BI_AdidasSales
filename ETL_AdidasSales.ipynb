```python
df.columns = [
    'retailer', 'retailer_id', 'invoice_date', 'region', 'state', 'city',
    'gender_type', 'product_category', 'price_per_unit', 'units_sold',
    'total_sales', 'operating_profit', 'operating_margin', 'sales_method'
]
df.info()
```

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'])

df['operating_profit'] = df['operating_profit'].str.replace(',', '.', regex=False).astype(float)
df['operating_margin'] = df['operating_margin'].str.replace(',', '.', regex=False).astype(float)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Fungsi deteksi outlier dengan IQR
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series < lower_bound) | (series > upper_bound)]

# Daftar kolom numerik (sesuaikan jika perlu)
numeric_cols = ['price_per_unit', 'units_sold', 'total_sales', 'operating_profit', 'operating_margin']

# Deteksi dan tampilkan outlier
for col in numeric_cols:
    outliers = detect_outliers(df[col])
    print(f"\nOutliers in '{col}':")
    print(outliers)
    print(f"Total outliers in '{col}': {len(outliers)}")

    # Scatter plot
    plt.figure(figsize=(6, 3))
    plt.scatter(df.index, df[col], label='Data', alpha=0.6)
    plt.scatter(outliers.index, outliers, color='red', label='Outliers')
    plt.title(f'Outliers in {col}')
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
df['year'] = df['invoice_date'].dt.year
df['month'] = df['invoice_date'].dt.month
df['quarter'] = df['invoice_date'].dt.to_period('Q')
df['day'] = df['invoice_date'].dt.day
df['weekday'] = df['invoice_date'].dt.day_name()
```

```python
categorical_cols = ['retailer', 'region', 'state', 'city', 'gender_type', 'product_category', 'sales_method']
for col in categorical_cols:
    df[col] = df[col].str.title().str.strip()
```

```python
negative_sales = df[df['total_sales'] < 0]
negative_price = df[df['price_per_unit']< 0]
negative_profit = df[df['operating_profit'] < 0]
negative_margin = df[df['operating_margin'] < 0]
negative_unit = df[df['units_sold'] < 0]

print(f"Negative sales records: {len(negative_sales)}")
print(f"Negative price records: {len(negative_price)}")
print(f"Negative profit records: {len(negative_profit)}")
print(f"Negative margin records: {len(negative_margin)}")
print(f"Negative unit records: {len(negative_unit)}")
```

```python
from sqlalchemy import create_engine, text
import pandas as pd

# Database connection
engine = create_engine('postgresql+psycopg2://postgres:nadia1624@localhost:5432/DW_Adidas')
```

```python
# Dimensi Retailer
dim_retailer = df[['retailer_id', 'retailer']].drop_duplicates().copy()
dim_retailer.columns = ['retailer_id', 'retailer_name']

# Dimensi Date
dim_date = df[['invoice_date']].drop_duplicates().copy()
dim_date['year'] = pd.to_datetime(dim_date['invoice_date']).dt.year
dim_date['month'] = pd.to_datetime(dim_date['invoice_date']).dt.month
dim_date['quarter'] = pd.to_datetime(dim_date['invoice_date']).dt.to_period('Q').astype(str)
dim_date['day'] = pd.to_datetime(dim_date['invoice_date']).dt.day
dim_date['weekday'] = pd.to_datetime(dim_date['invoice_date']).dt.day_name()
dim_date['date_id'] = range(1, len(dim_date) + 1)

# Dimensi Location
dim_location = df[['region', 'state', 'city']].drop_duplicates().copy()
dim_location['location_id'] = range(1, len(dim_location) + 1)

# Dimensi Product
dim_product = df[['product_category', 'price_per_unit']].drop_duplicates().copy()
dim_product['product_id'] = range(1, len(dim_product) + 1)

# Dimensi Gender
dim_gender = df[['gender_type']].drop_duplicates().copy()
dim_gender['gender_id'] = range(1, len(dim_gender) + 1)

# Dimensi Sales Method
dim_sales_method = df[['sales_method']].drop_duplicates().copy()
dim_sales_method['sales_method_id'] = range(1, len(dim_sales_method) + 1)
```

```python
fact_sales = df.merge(dim_retailer, on='retailer_id') \
               .merge(dim_date, on='invoice_date') \
               .merge(dim_location, on=['region', 'state', 'city']) \
               .merge(dim_product, on=['product_category', 'price_per_unit']) \
               .merge(dim_gender, on='gender_type') \
               .merge(dim_sales_method, on='sales_method')

fact_sales['sales_id'] = range(1, len(fact_sales) + 1)

fact_sales = fact_sales[['sales_id', 'retailer_id', 'date_id', 'location_id', 'product_id',
                         'sales_method_id', 'gender_id',
                         'units_sold', 'total_sales', 'operating_profit', 'operating_margin']]
```

```python
# Load into PostgreSQL
dim_retailer.to_sql('dim_retailer', engine, if_exists='append', index=False)
dim_date.to_sql('dim_date', engine, if_exists='append', index=False)
dim_location.to_sql('dim_location', engine, if_exists='append', index=False)
dim_product.to_sql('dim_product', engine, if_exists='append', index=False)
dim_gender.to_sql('dim_gender', engine, if_exists='append', index=False)
dim_sales_method.to_sql('dim_sales_method', engine, if_exists='append', index=False)
fact_sales.to_sql('fact_sales', engine, if_exists='append', index=False)

print("✅ Data berhasil dimuat ke PostgreSQL")
```
